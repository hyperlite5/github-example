```python
#import data from wiki page
from bs4 import BeautifulSoup
import requests
source = requests.get('http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source, 'lxml')
postal_data = soup.find('table')
import pandas as pd
pd.read_html(postal_data)
```

```python
#view data
postal_data
```

```python
#populate data into a dataframe
table_rows = postal_data.find_all('tr')

res = []

for tr in table_rows:

    row = tr.find_all('td')

    row = [tr.text.strip() for tr in row]

    if row:

        res.append(row)

df = pd.DataFrame(res, columns=["PostalCode", "Borough", "Neighbourhood"])

df #view new dataframe
```

```python
#remove not assigned from borough values
df1=df[df.Borough != 'Not assigned']
df1
```

```python
#replace items that have 'not assigned' in neighbourhood with the Bourough value
df1.loc[df1['Neighbourhood'] == ('Not assigned'), 'Neighbourhood'] = df1['Borough']
```

```python
#view data frame
df1
```

```python
#view shape of data frame
df1.shape
```

```python
#create a dataframe of latitude and longitude
geo_df = pd.read_csv('http://cocl.us/Geospatial_data')
geo_df

```

```python
#rename dataframe column to PostalCode
geo_df.columns = ['PostalCode','Latitude', 'Longitude']
geo_df
```

```python
#merged dataframe with PostalCode, Bourough, Neighbourhood, Latitude and Longitude
df2 = df1.merge(geo_df)
df2
```

```python
is_Etobicoke =  df2['Borough']=='Etobicoke'
is_Etobicoke
```

```python
#filter based on Etibicoke, Canada neighborhoods only
location_Etibicoke = df2[is_Etobicoke]
location_Etibicoke
```

```python
location_Etibicoke.shape
```

```python
address = 'Etobicoke, Canada'

geolocator = Nominatim(user_agent="Etobicoke_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Etobicoke are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Etobicoke using latitude and longitude values and showing how the neighborhoods cluster
Etobicoke2 = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, label in zip(location_Etibicoke['Latitude'], location_Etibicoke['Longitude'], location_Etibicoke['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(Etobicoke2)   
    
Etobicoke2
```
